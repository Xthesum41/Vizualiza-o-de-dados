```python
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")

penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data=penguins, hue='species')
```

```python
sns.histplot(x = 'species',hue='species',data=penguins)
```

```python
sns.histplot(x = 'body_mass_g', data=penguins, hue ='species', kde = True)
```

```python
sns.barplot(x='sex',y='bill_length_mm',data=penguins)
```

```python
sns.barplot(x='sex', y='bill_depth_mm',data=penguins,hue ='species')
```
